```python
"""
Main Entry Point
"""
from pathlib import Path
from typing import List
import argparse
from datetime import datetime

import pandas as pd # type: ignore

import src.fetch_data # type: ignore
import src.core_transform # type: ignore
import cdutils.pkey_sqlite # type: ignore
import cdutils.hhnbr # type: ignore


old_data = src.fetch_data.fetch_data_2024()
new_data = src.fetch_data.fetch_data_2025()

# # # Core transformation pipeline
old_data = src.core_transform.main_pipeline(old_data)
new_data = src.core_transform.main_pipeline(new_data)

# Raw data with pkey appended
old_data = cdutils.pkey_sqlite.add_pkey(old_data)
new_data = cdutils.pkey_sqlite.add_pkey(new_data)


 
```

```python
old_unique = old_data.groupby('branchname')['portfolio_key'].nunique().sort_values(ascending=False).reset_index()
```

```python
new_unique = new_data.groupby('branchname')['portfolio_key'].nunique().sort_values(ascending=False).reset_index()
```

```python
consolidation_unique = pd.merge(old_unique, new_unique, how='outer', on='branchname')
```

```python
consolidation_unique.info()
```

```python
consolidation_unique_schema = {'portfolio_key_x':'float'}
consolidation_unique = cdutils.input_cleansing.enforce_schema(consolidation_unique, consolidation_unique_schema)
```

```python
consolidation_unique = consolidation_unique.rename(columns={'portfolio_key_x': '2024_Q4','portfolio_key_y':'2025_Q1'}).copy()
```

```python
consolidation_unique['change'] = consolidation_unique['2025_Q1'] - consolidation_unique['2024_Q4']
```

```python
consolidation_unique['pct_change'] = (consolidation_unique['2025_Q1'] - consolidation_unique['2024_Q4']) / consolidation_unique['2024_Q4']
```
